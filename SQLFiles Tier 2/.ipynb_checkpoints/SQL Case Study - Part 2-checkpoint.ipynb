```python
import sqlite3
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)

    return conn
```

```python
def exec_query(conn, query):
    """
    Execute the query
    :param conn: the Connection object
    :param query: the Query string
    :return:
    """
    cur = conn.cursor()
    cur.execute(query)

    rows = cur.fetchall()

    for row in rows:
        print(row)
        
    print("\n")
```

```python
def main():
    database = r"sqlite_db_pythonsqlite.db"

    # create a database connection
    conn = create_connection(database)
    print("Created connection")
    
    with conn:
        print("Q10: Produce a list of facilities with a total revenue less than 1000. The output of facility name and total revenue, sorted by revenue. Remember that there's a different cost for guests and members! ")
        exec_query(conn, "SELECT name, SUM( CASE WHEN B.memid = 0 THEN guestcost* slots ELSE membercost * slots END) AS Revenue FROM Facilities F INNER JOIN Bookings AS B ON F.facid = B.facid  GROUP BY name HAVING Revenue < 1000 ORDER BY Revenue")

        print("Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order")
        exec_query(conn, "SELECT M.surname || \" \" || M.firstname AS Member, R.surname || \" \" || R.firstname AS RecommendedBy FROM Members AS M LEFT JOIN Members AS R ON M.recommendedby = R.memid ORDER BY Member")
        
        print("Q12: Find the facilities with their usage by member, but not guests")
        exec_query(conn, "SELECT F.name, M.surname || \" \" || M.firstname  AS Member, COUNT(B.bookid) AS FacilityUsage FROM  Bookings B INNER JOIN Facilities AS F ON B.facid = F.facid INNER JOIN Members AS M ON M.memid = B.memid AND B.memid <> 0 GROUP BY F.name, Member ORDER BY F.name, FacilityUsage DESC")
        
        print("Q13: Find the facilities usage by month, but not guests")
        exec_query(conn, "SELECT F.name,strftime('%m',B.starttime) AS Month ,COUNT(B.bookid) AS FacilityUsage FROM Bookings B INNER JOIN Facilities AS F ON B.facid = F.facid AND B.memid <> 0 GROUP BY F.name, Month ORDER BY F.name, Month ")
```

```python
main()
```
